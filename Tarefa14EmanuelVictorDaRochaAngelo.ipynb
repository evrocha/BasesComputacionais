```python
# EXERCICIO 02
data02 = pd.read_csv('https://raw.githubusercontent.com/evrocha/BasesComputacionais/master/dados/inflacaoMensal.csv')
# separando os elementos cujo o mes eh marco
data_marco = data02.loc[data02.Mês ==3] 
# mediana nos meses de marco
print(f"mediana = {data_marco['Inflação'].median():.2f}") # calculando a mediana da inflacao nos meses de marco
print(f"Média = {data_marco['Inflação'].mean():.2f}") # calculando a mediada inflacao nos meses de marco
print(f"Desvio-padrao = {data_marco['Inflação'].std():.2f}")  # calculando o desvio-padrao da inflacao nos meses de marco

plt.figure() # dando o plot do histograma referente a inflacao nos meses de marco
data_marco['Inflação'].hist()
plt.xlabel('Inflacao nos meses de marco')
plt.show()
```

```python
# EXERCICIO 04
import numpy as np
rng = np.random.default_rng()
x = rng.uniform(size= 100)
y = rng.uniform(size= 1000)
a = rng.uniform(size= 10000)
b = rng.uniform(size= 100000) 

plt.hist(x)
print(f"Média = {x.mean():.2f}") # calculando a media dos valores de x
plt.show()

plt.hist(y)
print(f"Média = {y.mean():.2f}") 
plt.show()

plt.hist(a)
print(f"Média = {a.mean():.2f}") 
plt.show()

plt.hist(b)
print(f"Média = {b.mean():.2f}") 
plt.show()
```
